```python
def obtener_escalas():
    '''
    Función que retorna una lista de diccionarios con la información de 
    las escalas de cada vuelo
    '''
        
    vuelos1 = driver.find_elements_by_xpath('//div[@class="leg selected ng-star-inserted"]')
    vuelos2 = driver.find_elements_by_xpath('//div[@class="leg ng-star-inserted"]')
    vuelos = vuelos1 + vuelos2
    print(f'Se encontraron {len(vuelos)} vuelos.')
    print('Iniciando scraping...')
    info = []
     
    for vuelo in vuelos:  
        info_escalas = {}
        # Hora de salida
        info_escalas['salida'] = vuelo.find_element_by_xpath('.//span[@class="hour departure ng-star-inserted"]').text
        # Hora de llegada
        info_escalas['llegada'] = vuelo.find_element_by_xpath('.//span[@class="hour arrival ng-star-inserted"]').text
        # Duracion
        info_escalas['duracion'] = vuelo.find_element_by_xpath('.//span[@class="time"]').text
        
        vuelo.find_element_by_xpath('.//a[@class="leg-details open-leg-details"]').click()

        segmentos = vuelo.find_elements_by_xpath('//div[@class="segment"]')        
        for segmento in segmentos:
            if segmento == segmentos[0]:               
                # Duración del vuelo
                info_escalas['duracion_vuelo'] = segmento.find_element_by_xpath('.//span[@class="time-duration ng-star-inserted"]//span[@class="time"]').text
                # Nombre de la aerolinea
                info_escalas['aerolinea'] = segmento.find_element_by_xpath('.//span[@class="airline-name"]').text
                # Numero del vuelo
                info_escalas['no_vuelo'] = segmento.find_element_by_xpath('.//strong[@class="flight-number"]').text
                # Clase
                info_escalas['clase'] = segmento.find_element_by_xpath('.//strong[@class="service-class"]').text

                aeropuertos = segmento.find_elements_by_xpath('.//p[@class="airport"]')
                for aeropuerto in aeropuertos:
                    if aeropuerto == aeropuertos[0]:
                        info_escalas['aeropuerto_1'] = aeropuerto.find_element_by_xpath('.//span[@class="airport-title"]').text
                        info_escalas['ciudad1'] = aeropuerto.find_element_by_xpath('.//span[@class="city-name"]').text
                        info_escalas['pais1'] = aeropuerto.find_element_by_xpath('.//span[@class="country-name"]').text
                        info_escalas['salida1'] = aeropuerto.find_element_by_xpath('.//span[@class="hour ng-star-inserted"]').text
                        info_escalas['dia1'] = aeropuerto.find_element_by_xpath('.//span[@class="day"]').text
                    elif aeropuerto == aeropuertos[1]:
                        info_escalas['aeropuerto_2'] = aeropuerto.find_element_by_xpath('.//span[@class="airport-title"]').text
                        info_escalas['ciudad2'] = aeropuerto.find_element_by_xpath('.//span[@class="city-name"]').text
                        info_escalas['pais2'] = aeropuerto.find_element_by_xpath('.//span[@class="country-name"]').text
                        info_escalas['salida2'] = aeropuerto.find_element_by_xpath('.//span[@class="hour ng-star-inserted"]').text
                        info_escalas['dia2'] = aeropuerto.find_element_by_xpath('.//span[@class="day"]').text

            elif segmento == segmentos[1]:
                #Duración del vuelo
                info_escalas['duracion_vuelo2'] = segmento.find_element_by_xpath('.//span[@class="time-duration ng-star-inserted"]//span[@class="time"]').text
                # Nombre de la aerolinea
                info_escalas['aerolinea2'] = segmento.find_element_by_xpath('.//span[@class="airline-name"]').text
                # Numero del vuelo
                info_escalas['no_vuelo2'] = segmento.find_element_by_xpath('.//strong[@class="flight-number"]').text
                # Clase
                info_escalas['clase2'] = segmento.find_element_by_xpath('.//strong[@class="service-class"]').text

                for aeropuerto in aeropuertos:
                    if aeropuerto == aeropuertos[0]:
                        info_escalas['aeropuerto3'] = aeropuerto.find_element_by_xpath('.//span[@class="airport-title"]').text
                        info_escalas['ciudad3'] = aeropuerto.find_element_by_xpath('.//span[@class="city-name"]').text
                        info_escalas['pais3'] = aeropuerto.find_element_by_xpath('.//span[@class="country-name"]').text
                        info_escalas['salida3'] = aeropuerto.find_element_by_xpath('.//span[@class="hour ng-star-inserted"]').text
                        info_escalas['dia3'] = aeropuerto.find_element_by_xpath('.//span[@class="day"]').text

                    elif aeropuerto == aeropuertos[1]:
                        info_escalas['aeropuerto_4'] = aeropuerto.find_element_by_xpath('.//span[@class="airport-title"]').text
                        info_escalas['ciudad4'] = aeropuerto.find_element_by_xpath('.//span[@class="city-name"]').text
                        info_escalas['pais4'] = aeropuerto.find_element_by_xpath('.//span[@class="country-name"]').text
                        info_escalas['salida4'] = aeropuerto.find_element_by_xpath('.//span[@class="hour ng-star-inserted"]').text
                        info_escalas['dia4'] = aeropuerto.find_element_by_xpath('.//span[@class="day"]').text

            else:
                None
                
        driver.find_element_by_xpath('.//div[@class="custom-dialog-close ng-star-inserted"]').click()
        info.append(info_escalas)
    return info
```

```python
def escala_espera(vuelo):
    espera = []
    tiempos_esperas = vuelo.find_elements_by_xpath('//div[@class="segment-change ng-star-inserted"]')
    for tiempo_espera in tiempos_esperas:
        if tiempo_espera == tiempos_esperas[0]:
            tiempo = tiempo_espera.find_element_by_xpath('.//strong[@class="qa-segment-change-time"]').text   
            dict_espera = {'Tiempo de espera':tiempo}
            espera.append(dict_espera)
        else:
            None
        
    return espera
```
