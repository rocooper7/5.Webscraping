```python
import re
is_well_formed_link = re.compile(r'^https?://.+/.+$') # https://example.com/hello
is_root_path = re.compile(r'^/.+$') # /some-text

def obtener_notas(soup):
    '''
    Función que recibe un objeto de BeautifulSoup de una página de una sección
    y devuelve una lista de URLs a las notas de esa sección
    '''
    lista_notas = []
       
    # Obtengo el listado de artículos
    article_list = soup.find('div', attrs={'class':'panel-panel grid-16 alpha'})
    for article in article_list.find_all('a'):
        if article:
            lista_notas.append(article.get('href'))
            
    lista_notas = list(set(link for link in lista_notas))
    
    #for link in range(len(lista_notas)):
        #lista_notas[link] = 'https://www.eluniversal.com.mx'  + lista_notas[link]
    
    lista = []
    for link in lista_notas:
        if is_well_formed_link.match(link):
            lista.append(link)            
        elif is_root_path.match(link):
            lista.append('https://www.eluniversal.com.mx' + link)
    
    return lista
```

```python
def obtener_info(s_nota):
    
    # Creamos un diccionario vacío para poblarlo con la información
    ret_dict = {}
    
    # Extraemos la fecha
    fecha = s_nota.find('div', attrs={'class':'fechap'})
    if fecha:
        ret_dict['fecha'] = fecha.text
    else:
        ret_dict['fecha'] = None
    
    # Extraemos el título
    titulo = s_nota.find('div', attrs={'class':'panel-pane pane-node-title'})
    if titulo:
        ret_dict['titulo'] = titulo.text
    else:
        ret_dict['titulo'] = None

    # Extraemos la volanta
    volanta = s_nota.find('div', attrs={'class':'field field-name-field-resumen field-type-text-long field-label-hidden'})
    if volanta:
        ret_dict['volanta'] = volanta.get_text()
    else:
        ret_dict['volanta'] = None
    
    autor = s_nota.find('div', attrs={'field-item even'})
    if autor:
        ret_dict['autor'] = autor.get_text()
    else:
        ret_dict['autor'] = None
        
    # Extraemos el cuerpo de la nota
    cuerpo = s_nota.find('div', attrs={'field field-name-body field-type-text-with-summary field-label-hidden'})
    if cuerpo:
        ret_dict['texto'] = cuerpo.get_text()
    else:
        ret_dict['texto'] = None
    
    return ret_dict
```

```python
def scrape_nota(url):
    try:
        nota = requests.get(url)
    except Exception as e:
        print('Error scrapeando URL', url)
        print(e)
        return None
    
    if nota.status_code !=200:
        print(f'Error obteniendo nota {url}')
        print(f'stuts Code = {nota.status_code}')
        return None
    
    s_nota = BeautifulSoup(nota.text, 'lxml')
    
    ret_dict = obtener_info(s_nota)
    ret_dict['url'] = url
    
    return ret_dict
```

```python
notas = []
for link in links:
    try:
        r = requests.get(link)
        if r.status_code == 200:
            soup = BeautifulSoup(r.text, 'lxml')
            notas.extend(obtener_notas(soup))
        else:
            print('No se pudo obtener la sección', link)
    except:
        print('No se pudo obtener la sección', link)
```

```python
data = []
for i, nota in enumerate(notas):
    print(f'Scrapeando nota{i}/{len(notas)}')
    data.append(scrape_nota(nota))
```
