```python
from bs4 import BeautifulSoup
s = BeautifulSoup(r.text,'lxml')
s
```

```python
print(s.prettify())
```

```python
secciones1 = s.find('ul', attrs={'class':'navegacion-listado'}).find_all('li')
secciones1
```

```python
secciones2= s.find('div',{'class':'col-lg-3 col-md-3 col-xs-6 col-2'}).find_all('li')
secciones2
```

```python
secciones3 =s.find('div',{'class':'col-lg-6 col-md-6 col-xs-12 col-3'}).find_all('li')
secciones3
```

```python
secciones1 = secciones1 + secciones2 + secciones3 
secciones1
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones1]
links_secciones
```

```python
for link in range(len(links_secciones)):
    links_secciones[link] = 'https://www.sdpnoticias.com' + links_secciones[link]
links_secciones
```
